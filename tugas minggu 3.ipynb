```python
unik_data = df.drop_duplicates(subset=['pizza_name_id'])

# Membuat bar chart
plt.bar(unik_data['pizza_category'], unik_data['pizza_size'])
plt.xlabel('Jenis Pizza')
plt.ylabel('Jumlah Order')
plt.title('Jumlah Order Pizza Unik')
plt.show()
```

```python
import seaborn as sns


# Membuat box plot untuk harga pizza
plt.figure(figsize=(10,6))
sns.boxplot(x='pizza_category', y='unit_price', data=df)
plt.title('Box Plot Harga Pizza')
plt.xlabel('Kategori Pizza')
plt.ylabel('Harga')
plt.show()
```

```python
corr_matrix = df[['unit_price', 'quantity']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Heatmap Korelasi')
plt.show()
```
